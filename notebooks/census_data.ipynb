```python
census_fields = {
    'B19013_001E': 'median_hh_income',
    'B02001_001E': 'total_pop_race',
    'B02001_002E': 'white_only_pop',
    'B02001_003E': 'black_only_pop',
    'B02001_008E': 'multiracial_pop',
    'B03001_003E': 'hispanic_pop',
    'B07013_001E': 'total_pop_tenure',
    'B07013_003E': 'total_renter_pop',
    'B07013_006E': 'renter_non_mover',
    'B07013_009E': 'renter_same_county_movers',
    'B07013_012E': 'renter_same_state_movers',
    'B07013_015E': 'renter_diff_state_movers',
    'B07013_018E': 'renter_diff_country_movers',
    'B25002_002E': 'total_occupied_units',
    'B25003_003E': 'total_rental_tenure',
    'B25038_010E': 'rental_movein_2005_later',
    'B25038_011E': 'rental_movein_2000_2004',
    'B25038_012E': 'rental_movein_1990_1999',
    'B25038_013E': 'rental_movein_1980_1989',
    'B25038_014E': 'rental_movein_1970_1979',
    'B25038_015E': 'rental_movein_1969_earlier',
    'B25039_002E': 'median_movein_yr_owners',
    'B25039_003E': 'median_movein_yr_renters'
}
```

```python
data = c.acs5.state_county_tract(fields=list(census_fields.keys()), state_fips='06', county_fips='075', tract="*")
```

```python
census_df = pd.DataFrame(data).rename(columns=census_fields)
```

```python
census_df['renter_movein_1969_earlier'] + census_df['renter_movein_1970_1979'] + census_df['renter_movein_1980_1989'] + \
 census_df['renter_movein_1990_1999'] + census_df['renter_movein_2000_2004'] + \
census_df['renter_movein_2005_later']
```
